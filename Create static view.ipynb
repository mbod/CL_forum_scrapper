```python
import os
import pickle
import re
import random
import time
import requests
from bs4 import BeautifulSoup
```

```python
threads = [os.path.join('data/threads',f) for f in os.listdir('data/threads') if f.endswith('.html')]
```

```python
post_url = 'https://forums.craigslist.org/?act=Q&ID={}'
```

```python
post2thread = pickle.load(open('data/post2thread.pkl','rb'))
```

```python
post2thread
```

```python
def process_thread(thread_file):
    print('Processing', thread_file)
    thread=BeautifulSoup(open(thread_file).read())
    p=thread.find('p')
    p.string=p.text.replace(p.text[:p.text.index('thread')],'')
    thread_id = thread_file.split('/')[-1].replace('.html','')
    for a in thread.select('div.threadline > a.title'):
        post_id = re.search('ID=(\d+)',a.attrs['href']).group(1)
        if post_id not in post2thread:
            post2thread[post_id] = thread_id
        a.attrs['href']=re.sub('.*ID=(\d+)','../posts/\\1.html', a.attrs['href'])
        if not os.path.exists('data/posts/{}.html'.format(post_id)):
            print('no post html saved for', post_id)
            presp = requests.get(post_url.format(post_id))
            pdoc = BeautifulSoup(presp.text)
            post_div = pdoc.select('div.displaypost > div.normal')[0]
            atag=post_div.find('a')
            atag.name='span'
            with open('data/posts/{}.html'.format(post_id),'w') as pout:
                pout.write(post_div.prettify())
            time.sleep(random.randint(1,2))
            
    with open('view/threads/{}'.format(thread_file.split('/')[-1]),'w') as tout:
        tout.write(thread.prettify())
```

```python
for thread in threads:
    process_thread(thread)
```
